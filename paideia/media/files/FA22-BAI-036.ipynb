# NLTK - Natural Language Toolkit

Python Programming can be used to process text data for the requirements in various textual data analysis. A very important area of application of such text processing ability of python is for NLP (Natural Language Processing). NLP is used in search engines, newspaper feed analysis and more recently for voice -based applications like Siri and Alexa. Python's Natural Language Toolkit (NLTK) is a group of libraries that can be used for creating such Text Processing systems.

```python
text = """The feature on the new Pixel 7 is not as good Apple Inc’s Face ID unlocking mechanism,
as it can struggle in low light and is more vulnerable to being spoofed. In addition, Google has
said it is not secure enough to enable signing into apps or making payments.

The return comes after Google became stricter about launching products with facial recognition,
in part due to questions about its performance on darker skin. The company took time to review its
approach to training and testing facial recognition since the previous Pixel with the capability
launched in 2019, one of the sources said.

Google declined to comment on several specific questions about its history with face unlock. It
said generally, “Thanks to advanced machine learning models for face recognition, Pixel 7 and
Pixel 7 Pro feature Face Unlock, but we’re doing it a little differently.”"""
```

```python
import nltk
nltk.download('punkt')
import matplotlib.pyplot as plt
```

```python
# Create Uni-grams from the above text and then plot the frequency
# YOUR CODE HERE

from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

tokens = word_tokenize(text)
fdist = FreqDist(tokens)

fdist.plot(30, cumulative=False)
plt.title('Frequency Distribution of Uni-grams')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create Bi-grams from the above text and then plot the frequency
# YOUR CODE HERE
from nltk import word_tokenize
from nltk.util import ngrams


# Tokenize the text
tokens = word_tokenize(text.lower())

# Generate bi-grams
bi_grams = list(ngrams(tokens, 2))

# Calculate frequency distribution
freq_dist = nltk.FreqDist(bi_grams)

# Plotting
plt.figure(figsize=(10, 6))
freq_dist.plot(20, title='Top 20 Bi-grams Frequency')
plt.xlabel('Bi-gram')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create Tri-grams from the above text and then plot the frequency
# YOUR CODE HERE
tokens = word_tokenize(text.lower())

tri_grams = list(ngrams(tokens, 3))

freq_dist = nltk.FreqDist(tri_grams)

plt.figure(figsize=(10, 6))
freq_dist.plot(20, title='Top 20 Tri-grams Frequency')
plt.xlabel('Tri-gram')
plt.ylabel('Frequency')
plt.show()
```

```python
# Display the Uni-gram frequency distribution using matplotlib. Add proper x and y labels.
#YOUR CODE HERE
tokens = word_tokenize(text.lower())

freq_dist = nltk.FreqDist(tokens)

plt.figure(figsize=(10, 6))
freq_dist.plot(20, title='Top 20 Uni-grams Frequency')
plt.xlabel('Uni-gram')
plt.ylabel('Frequency')
plt.show()
```

```python
import nltk
from nltk.corpus import stopwords

# Download stopwords list if not already downloaded
nltk.download('stopwords')

def remove_stopwords(word_list):
    stop_words = set(stopwords.words('english'))
    processed_word_list = [word for word in word_list if word.lower() not in stop_words]
    return processed_word_list

# Example usage:
# Assuming token_text is already defined
token_text = ["The", "quick", "brown", "fox", "jumps", "over", "the", "lazy", "dog"]
removed_stopwords = remove_stopwords(token_text)
print(removed_stopwords)
```

```python

def remove_punctuation(text):
    import string
    translator = str.maketrans('', '', string.punctuation)
    return text.translate(translator)

# Example usage:
text = "Hello, world! This is a sample text with punctuation."
cleaned_text = remove_punctuation(text)
print(cleaned_text)

# Now create a frequency plot:
tokens = word_tokenize(cleaned_text)
fdist = FreqDist(tokens)

plt.figure(figsize=(10, 6))
fdist.plot(20, title='Top 20 Uni-grams Frequency')
plt.xlabel('Uni-gram')
plt.ylabel('Frequency')
plt.show()

```
